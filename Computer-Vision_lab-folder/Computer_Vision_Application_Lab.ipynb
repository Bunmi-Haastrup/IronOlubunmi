## Ethical issues with Computer Vision Applications

The computer vision technologies are being used to capture more detailed and personalised information in a specific point of time. However, concerns on ethics of computer vision should not be neglected due to its significant impact on people's lives, culture and society both positively and negatively. As the technology evolve, there should be public awareness and adequate information on the pros and cons. 

The right to privacy should be a point of major concern when using facial recognition algorithm, there should be a consideration of what the society wants.

The data should not be collected or used without full consent of the people involved. Full details about the data collection must be provided before data collection. For example, it is illegal for a commercial firm to use a person's biometric data without their consent and adequate information about the process. 

Detailed information should be provided, researcher's should not take advantage of the vulnerability of the people. A lot of people fall victim of this without having enough details of the purpose of the data or research to have given full consent.
