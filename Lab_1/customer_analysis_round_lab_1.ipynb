### DATA CLEANING

```python
# Drop the duplicates from the dataframe
new_df = new_df.drop_duplicates()
```

```python
new_df.isna().sum() # Check for missing values
```

```python
# drop the education and number of open complaints columns
to_drop = ['education','number_of_open_complaints']
new_df.drop(to_drop,axis = 1,inplace =True)
```

```python
new_df
```

```python
new_df['customer_lifetime_value'].isnull().sum()
```

```python
def clean(colm):
    return round(float(str(colm).replace('%', ''))*100,3)
```

```python
new_df['customer_lifetime_value'] = new_df['customer_lifetime_value'].apply(clean)
```

```python
new_df[['customer_lifetime_value']].astype(float)
```

```python
new_df[['customer_lifetime_value']].dropna(inplace = True)
```

```python
new_df.loc[new_df['st'].isna(),['st','gender','state']]
```

since 'st' has a lot of missing values then it is advisable to drop the column and stick with the other state column

```python
new_df.drop('st',axis = 1,inplace = True)
```

```python
new_df = new_df.dropna()
```
