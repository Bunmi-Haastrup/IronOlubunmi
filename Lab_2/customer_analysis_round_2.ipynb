```python
dtf1 = pd.read_csv('file1.csv')
dtf2 = pd.read_csv('file2.csv')
dtf3 = pd.read_csv('file3.csv')
```

```python
cols = []
for col in dtf1.columns:
    x = col.lower().replace(' ','_')
    cols.append(x)
dtf1.columns = cols
```

```python
colum = []
for col in dtf2.columns:
    y = col.lower().replace(' ','_')
    colum.append(y)
dtf2.columns = colum
```

```python
column = []
for col in dtf3.columns:
    z = col.lower().replace(' ','_')
    column.append(z)
dtf3.columns = column
```

```python
df_new = pd.concat([dtf1,dtf2,dtf3])
```

```python
df_new = df_new.reset_index(drop=True)
```
